```python
from collections import Counter
```

```python
word_count = Counter()
for word in re.finditer(PAT, text):
    word = word.captures()[0]
    word_count[tuple(word.encode())] += 1
print(word_count)
```

```python
def get_word_count(text):
    word_count = Counter()
    for word in re.finditer(PAT, text):
        word = word.captures()[0]
        word_count[tuple(word.encode("utf-8"))] += 1
        # word_count[word.encode("utf-8")] += 1

    return word_count
```

```python
pair_counts = Counter()
for word, freq in word_count.items(): 
    for p in zip(word, word[1:]):
        pair_counts[p] += freq
print(pair_counts)
```

```python
def get_pair_count(word_count):
    pair_count = Counter()
    for word, freq in word_count.items(): 
        for p in zip(word, word[1:]):
            pair_count[p] += freq

    return pair_count
```

```python
top_pair = max(pair_counts, key=pair_counts.get)
top_pair
```

```python
def merge(word_count, pair_count, top_pair, new_byte_idx):
    new_word_count = Counter()
    for word, freq in word_count.items(): # word_count -> dict[tuple[bytes], int]
        new_word = []
        i = 1 # `i` denotes index of end of pair
        idxs = []

        new_word.append(word[0])
        while i < len(word):
            if (word[i-1], word[i]) == top_pair:
                idxs.append(i)
                new_word.pop()
                new_word.append(new_byte_idx)
                pair_count[top_pair] -= freq # decrement top_pair's count
                if i != 1:
                    pair_count[(word[i-2], word[i-1])] -= freq # decrement left neighbour
                if i+1 < len(word):
                    pair_count[(word[i], word[i+1])] -= freq # decrement right neighbour
                    new_word.append(word[i+1])

                i += 2
            else:
                new_word.append(word[i])
                i += 1

        new_word = tuple(new_word)
        if idxs:
            # create and insert the new tuple
            new_word_count[new_word] = freq
            # add new neighbours' counts
            for idx in idxs:
                if idx != 1:
                    pair_count[(word[idx-2], new_byte_idx)] += freq # left neighbour
                if idx+1 < len(word):
                    pair_count[(new_byte_idx, word[idx+1])] += freq # right neighbour
        else:
            new_word_count[word] = freq

    return new_word_count
```

```python
word_count = get_word_count(text)
pair_count = get_pair_count(word_count)
merges = {}
vocab = {idx: bytes([idx]) for idx in range(256)}

vocab_sz = 300
num_merges = vocab_sz - 256
for i in range(num_merges):
    new_byte_idx = 256 + i
    top_pair = max(pair_count, key=pair_count.get)
    if pair_count[top_pair] == 0: break
    word_count = merge(word_count, pair_count, top_pair, new_byte_idx)
    print(pair_count[top_pair])
    merges[top_pair] = new_byte_idx
    vocab[new_byte_idx] = top_pair
```

```python
# sanity check
for k,v in get_pair_count(word_count).items():
    if v != pair_count[k]:
        print(f"k: {k}; actual v: {v}; pair_count v: {pair_count[k]}")
```
